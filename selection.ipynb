```python
import pandas as pd
import os 
current_dir = os.getcwd()
import shutil
folder = os.path.join(current_dir, "selected_data")
if os.path.isdir(folder):
    shutil.rmtree(folder)
    os.mkdir(folder)
```

```python
train = pd.read_csv("Dataset\\train.tsv", sep='\t', names=['id', 'label', 'statement', 'subject(s)', 'speaker', 'speaker\'s job title', 'state info', 'party affiliation', 'credit history count', 'true count', 'false count', 'half true count', 'most true count', 'pants on fire count', 'context'], header=None)
```

```python
train.head()
```

```python
train.info()
```

```python
train = train[['label', 'statement']]
```

```python
train.head()
```

```python
train.to_csv(os.path.join(folder, 'train.csv'), index=False)
```
